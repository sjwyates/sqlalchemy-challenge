# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Get me some column names
inspector = inspect(engine)
m_columns = [col['name'] for col in inspector.get_columns('measurement')]
s_columns = [col['name'] for col in inspector.get_columns('station')]
print(f"measurement columns: {m_columns}")
print(f"station columns: {s_columns}")
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from the last data point in the database
last_date = session.query(Measurement.date)\
    .order_by(Measurement.date.desc()).first()[0]\
    .split('-')
last_date = dt.date(*[int(str) for str in last_date])
year_b4 = dt.date(last_date.year - 1, last_date.month, last_date.day)

# Perform a query to retrieve the data and precipitation scores
last_12_mos = session.query(Measurement.date, Measurement.prcp)\
    .filter(func.strftime('%Y-%m-%d', Measurement.date) >= year_b4)\
    .all()

# Save the query results as a Pandas DataFrame and set the index to the date column
meas_df = pd.DataFrame(last_12_mos, columns=['Date', 'Precipitation'])

# Sort the dataframe by date
meas_df = meas_df\
    .sort_values(by=['Date'])\
    .dropna()\
    .reset_index(drop=True)

# Use Pandas Plotting with Matplotlib to plot the data
meas_df.plot(x='Date', y='Precipitation', figsize=(10, 6), linewidth=1)
plt.title('Precipitation Analysis', fontsize=16)
plt.legend(prop={'size': 12})
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Inches', fontsize=12)
plt.tight_layout()
plt.savefig('Images/prcp-analysis.png')
plt.plot()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
meas_df.describe().round(2)
```

```python
# Design a query to show how many stations are available in this dataset?
active_stations = session.query(Measurement.station, func.count(Measurement.station))\
    .group_by(Measurement.station)
num_stations = active_stations.count()
print(f"Number of Stations: {num_stations}")
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
most_active_stations = active_stations.order_by(func.count(Measurement.station).desc()).all()
print(most_active_stations)
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
most_active_station = session.query(Measurement.tobs)\
    .filter(Measurement.station == most_active_stations[0][0])

lowest_temp = most_active_station.order_by(Measurement.tobs).first()[0]
print(f"Lowest Temp: {lowest_temp}")
highest_temp = most_active_station.order_by(Measurement.tobs.desc()).first()[0]
print(f"Highest Temp: {highest_temp}")
average_temp = session.query(func.avg(Measurement.tobs)).first()[0]
print(f"Average Temp: {round(average_temp, 1)}")
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
most_temp_12_mos = most_active_station\
    .filter(func.strftime('%Y-%m-%d', Measurement.date) >= year_b4)\
    .all()

most_active_df = pd.DataFrame(most_temp_12_mos, columns=['Temperature'])

most_active_df.plot.hist(figsize=(8, 6))
plt.title(f'Station {most_active_stations[0][0]} Temperature Observations', fontsize=14)
plt.legend(prop={'size': 12})
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Temperature Range', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.savefig('Images/prcp-analysis.png')
plt.plot()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```
